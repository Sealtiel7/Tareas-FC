```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import calendar
import datetime

#Este programa está destinado a leer los datos de la CONAGUA y determinar
#la información de cada mes en diferentes parámetros.

#os.chdir('/Users/Usuario/Python/Física Computacional/Activida_3/Actividad_4')

#Este bloque determina los elementos nulos de la tabla de datos.
missing=['NULO']
df = pd.read_csv('fuentes.txt', sep='\s+', na_values=missing)
pd.isnull(df)
df['FECHA'] = pd.to_datetime(df['FECHA'])

df.dropna()
```

```python
df['Año'] = df['FECHA'].dt.year
df['Mes'] = df['FECHA'].dt.month 
print(df.FECHA)
```

```python
n=len(df['Año'].unique())
print("El Archivo de texto contiene",n , "años en datos")
```

```python
tot=0.0
for i in range(12):
    PrecipM = df['PRECIP'][df['Mes']==[i+1]].sum()/n
    tot=tot+PrecipM
    print("Mes", i+1,":", np.round(PrecipM, decimals=2), "mm", ", Acumulada:", 
          np.round(tot, decimals=2), "mm")
    #aprecip=[i+1,PrecipM]
    #print(aprecip)
    plt.bar(i+1, PrecipM,.5)
    plt.savefig("barmen")
   
    
    
```

```python
for i in range(1954,1986):
    PrecipA = df['PRECIP'][df['Año']==[i+1]].sum()
    #print("Año", i+1,":", PrecipA, "mm")
    plt.bar(i+1,PrecipA,.5)
    plt.savefig("baran")
     
```

```python
co = df[["TMAX","TMIN"]]
fig = plt.plot(df.Año, co)

plt.grid(True)
#plt.show()
plt.savefig("evo")    
    
```

```python
#Digrma de caja de las temperaturas mínimas promedio.
for i in range(12):
    TminProm = df['TMIN'][df['Mes']==[i+1]].mean()
    
    #print("Mes", i+1,":", np.round(TmaxProm, decimals=2), "GRados")
    #aprecip=[i+1,PrecipM]
    #print(aprecip)
    arra=[i+1,TminProm]
    print(arra)
    plt.boxplot(arra)
    plt.savefig("caja")
    

```
